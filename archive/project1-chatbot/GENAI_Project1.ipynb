```python
!pip install transformers torch gradio --quiet

from transformers import pipeline
import gradio as gr

# Better chat model - still small enough for free Colab
generator = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    torch_dtype="float16"   # Saves memory
)

def chatbot(message):
    try:
        if not message or len(message.strip()) < 3:
            return "Please ask a proper question!"

        # Good prompt format for this model
        prompt = f"<|user|>\n{message}\n<|assistant|>\n"

        result = generator(
            prompt,
            max_new_tokens=200,
            temperature=0.7,
            do_sample=True,
            top_p=0.9,
            pad_token_id=generator.tokenizer.eos_token_id
        )

        response = result[0]['generated_text']

        # Clean output
        if "<|assistant|>" in response:
            response = response.split("<|assistant|>")[-1].strip()

        return response.strip() if response else "I couldn't generate a good response. Try rephrasing!"

    except Exception as e:
        return f"Error: {str(e)[:100]}... Try again with a simpler question."

# Launch
gr.Interface(
    fn=chatbot,
    inputs=gr.Textbox(lines=3, placeholder="Ask me anything about GenAI, energy investments, or delivery projects..."),
    outputs=gr.Textbox(label="AI Response"),
    title="🚀 My First GenAI Chatbot (TinyLlama 1.1B)",
    description="Improved model - much better than distilgpt2",
    examples=[
        ["Explain generative AI simply"],
        ["Give ideas for energy investment analysis using AI"],
        ["Summarize my package delivery prediction project"]
    ]
).launch(share=True)
```
