```python
# Load models
sentiment_pipeline = pipeline("sentiment-analysis")

summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6"
)

print("✅ Both models loaded successfully!")

def analyze_text(text):
    if not text or len(text.strip()) < 10:
        return "Please enter meaningful text!", "Text too short for summarization."

    # Sentiment
    sentiment_result = sentiment_pipeline(text)[0]
    sentiment_output = f"{sentiment_result['label']} ({sentiment_result['score']:.2%} confidence)"

    # Summarization
    try:
        summary_result = summarizer(text, max_length=130, min_length=30, truncation=True)
        summary = summary_result[0]['summary_text']
    except Exception as e:
        summary = f"Summary error (try shorter text): {str(e)[:80]}"

    return sentiment_output, summary

# Launch App
gr.Interface(
    fn=analyze_text,
    inputs=gr.Textbox(lines=12, placeholder="Paste your energy report, project notes, or reviews here..."),
    outputs=[
        gr.Textbox(label="📊 Sentiment Analysis"),
        gr.Textbox(label="📝 Summary")
    ],
    title="🚀 Project 2: Sentiment Analysis + Text Summarizer",
    description="Free Hugging Face models",
    examples=[
        ["I love how this new energy investment strategy is performing. The returns are excellent and sustainable."],
        ["The package delivery prediction model needs major improvement. Accuracy is low and users are frustrated."]
    ]
).launch(share=True)
```
