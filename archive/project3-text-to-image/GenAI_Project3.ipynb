```python
!pip install diffusers transformers gradio accelerate --quiet

import gradio as gr
from diffusers import StableDiffusionPipeline
import torch

# Load the model
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda")

def generate_image(prompt):
    if not prompt:
        return None
    image = pipe(prompt, num_inference_steps=30, guidance_scale=7.5).images[0]
    return image

gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(lines=2, placeholder="A futuristic energy efficient city, cyberpunk style..."),
    outputs=gr.Image(label="Generated Image"),
    title="🚀 Project 3: Text-to-Image Generator (Stable Diffusion)",
    description="Generate images from text using free Stable Diffusion",
    examples=[
        ["Futuristic delivery drone flying over green energy fields"],
        ["Professional energy investment dashboard visualization"],
        ["Package delivery robot in a smart city"]
    ]
).launch(share=True)
```
